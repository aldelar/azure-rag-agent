```python
# generate chunks from markdown
def generate_chunks_from_markdown(markdown):
	raw_chunks = markdown.split("## ")
	raw_chunks = list(filter(None, raw_chunks))
	chunks = []
	for i, raw_chunk in enumerate(raw_chunks):
		try:
			title, content = raw_chunk.split("\n", 1)
			content = title + "\n" + content
		except:
			title = ""
			content = raw_chunk
		chunks.append({ "index": i, "title": title, "content": content})
	return chunks
```

```python
chunks = generate_chunks_from_markdown(markdown)
```

```python
for chunk in chunks:
	print(chunk)
```
